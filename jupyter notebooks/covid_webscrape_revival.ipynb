## POPULATION WEBScrape 

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup 
import matplotlib.pyplot as plt
import pandas as pd
import dask.dataframe as dd
```

```python
browser = Browser('chrome')
```

```python
# Visit the website
# reference https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://www.worldometers.info/coronavirus/#countries"
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')
rows = table.find_all('tr')

```

```python
# Create an empty list to store each row of data
country_data = []


# Loop through the scraped rows to create a list of rows
for x in rows[1:]:
        td = x.find_all('td')
        row = [col.text for col in td]
        country_data.append(row)
country_data 

```

```python
country_data  = pd.DataFrame(country_data , columns = ['id','Country', 'Total_Cases', 'New_Cases', 'Total_Deaths', 'Total_Recovered', 'New_Recovered', 'Active', 'Serious', 'trash10', 'trash11', 'trash12','trash13','trash14','trash15','trash16','trash17','trash18','trash19','trash20','trash21','trash22'])


```
