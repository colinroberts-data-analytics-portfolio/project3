## POPULATION WEBScrape 

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup 
import matplotlib.pyplot as plt
import pandas as pd
import dask.dataframe as dd
```

```python
browser = Browser('chrome')
```

```python
# Visit the website
# reference https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://www.worldometers.info/world-population/population-by-country/"
browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')
rows = table.find_all('tr')

```

```python
# Create an empty list to store each row of data
pop_data = []


# Loop through the scraped rows to create a list of rows
for x in rows[1:]:
        td = x.find_all('td')
        row = [col.text for col in td]
        pop_data.append(row)
pop_data 

```

```python
pop_data = pd.DataFrame(pop_data, columns = ['id','Country', 'Population', 'Yearly_Change', 'Net_Change', 'Density', 'Land_Area', 'Migrant', 'Fertility_Rate', 'Medium_Age', 'Urban_Population', 'World_share'])


```

```python
print(pop_data.dtypes)
```

```python
print(pop_data)
```

## replace commas

```python

pop_data['Country'] = pop_data['Country'].astype(str)

# Convert 'Population', 'Density', and 'Land_Area' to strings
pop_data['Population'] = pop_data['Population'].astype(str)
pop_data['Density'] = pop_data['Density'].astype(str)
pop_data['Land_Area'] = pop_data['Land_Area'].astype(str)

# Remove commas and convert the columns to integers
pop_data['Population'] = pop_data['Population'].str.replace(',', '').astype(int)
pop_data['Density'] = pop_data['Density'].str.replace(',', '').astype(int)
pop_data['Land_Area'] = pop_data['Land_Area'].str.replace(',', '').astype(int)

# Print the cleaned DataFrame and check the data types
print(pop_data)
print(pop_data.dtypes)

```
